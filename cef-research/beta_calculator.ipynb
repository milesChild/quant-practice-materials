```python
"""
Formula for calculating the beta between two assets.

Using for calculating the beta between PHYSX and GLD.
"""
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Get price data for PHYSX and GLD from yfinance
PHYSX = yf.download('PHYS', start='2015-01-01', end='2020-01-01')
GLD = yf.download('GLD', start='2015-01-01', end='2020-01-01')
```

```python
# Show a time series of physx
PHYSX['Adj Close'].plot(figsize=(10, 7))
plt.show()
```

```python
# Show a time series of gld
GLD['Adj Close'].plot(figsize=(10, 7))
plt.show()
```

```python
# Plot the daily percent changes against eachother as a time series
(PHYSX['Adj Close'].pct_change()).plot()
(GLD['Adj Close'].pct_change()).plot()
plt.legend(['PHYSX', 'GLD'])
plt.show()
```

```python
# Calculate the beta between the two assets
covariance = PHYSX['Adj Close'].pct_change().cov(GLD['Adj Close'].pct_change())
variance = GLD['Adj Close'].pct_change().var()
PHYSX_beta = covariance / variance
print(PHYSX_beta)
```
