```python
import yfinance as yf
import pandas as pd
import numpy as np
import time
```

```python
current_date = "2023-01-20"
one_yr_ago = "2018-01-01"
usa_nav = yf.download("XUSAX", start=one_yr_ago, end=current_date)
usa_nav.head(15)
```

```python
# Remove everything except the Adj Close column
usa_nav = usa_nav.drop(columns=['Open', 'High', 'Low', 'Close', 'Volume'])
usa_nav.head(15)
```

```python
# Round all the prices to 4 decimal places
usa_nav = usa_nav.round(4)
usa_nav.head(15)
```

```python
# Conert the dataframe to a dictionary
usa_nav_dict = usa_nav.to_dict()
print(usa_nav_dict)
```

```python
xusax = pd.DataFrame(usa_nav_dict)
xusax.head(15)
```
